## 1\. Preço do diamante por tipo de corte

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
# gráfico do exercício 1
diamonds = data[['cut','price','color']].groupby('cut').agg('mean').reset_index()
diamonds.head(30)

grafico_barras = sns.barplot(data=diamonds,x="cut", y="price")
grafico_barras.set(title="Gráfico com a média de valores dos diamantes por corte", xlabel= 'Tipo de corte', ylabel="Valores")
#grafico_barras.figure(figsize=(12,8))
```

**Insight do gráfico 1**: ...

Analisando o gráfico acima pode-se notar que a média de valor altera de acordo com o tipo de corte aplicado no diamante, como por exemplo o corte "Premium" tem uma média de valor acima do corte "Ideal"

## 2\. Número de passageiros em dezembro por ano

```

import seaborn as sns

data = sns.load_dataset("flights")
data.head()
```

```
# gráfico do exercício 2

flights = data.query('month == "Dec"')
flights.head()

with sns.axes_style('darkgrid'):
  grafico = sns.lineplot(data=flights, x='year', y='passengers', color='red')
  grafico.set(title="Número de passageiros anuais somente no mês de dezembro", xlabel='Ano', ylabel='Passageiros')
```

**Insight do gráfico 2**: ...

Analisando o gráfico acima, nota-se que o número de passagerios anuais somente no mês de dezembro sobre de acordo com o passar dos anos, somente entre 1957 e 1958 que pode-se notar que a diferença de valores não foi significante.

## 3\. Numero de passageiros por mês entre 1949 e 1959

```

import seaborn as sns

data = sns.load_dataset("flights")
data.head()
```

```
# gráfico do exercício 3

flights = data.query("1949 <= year < 1960")

with sns.axes_style('whitegrid'):

  grafico = sns.lineplot(data=flights, x='month', y='passengers', hue='year')
  grafico.set(title='Passageiros por mês por ano', xlabel='Mês', ylabel='Passageiros');
  grafico.get_legend().set_title("Ano");
```

**Insight do gráfico 3**: ...

Nota-se que nos anos de 1958 e 1957 a diferença de passageiros foi de zero no mês de Abril, ou seja, a mesma quantidade no mesmo mês no período de um ano de diferença, e isso ocorre tmabém no mês de Setembro e provavelemnte em Dezembro, no mesmo período de um ano.

Em 1960 a companhia teve um grande aumento no número de passageiros durante o ano todo e em nenhum momento esteve igual o ano anterior.
